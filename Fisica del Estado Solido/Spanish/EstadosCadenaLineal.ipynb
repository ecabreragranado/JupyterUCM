Algunos óxidos de metales de transición se comportan como conductores
casi-unidimensionales, como es el caso del compuesto K$_{0.3}$MoO$_{3}$. Para 
estudiar sus propiedades electrónicas se formula un sencillo modelo basado en
una red atómica unidimensional, donde se considera un único orbital para cada
átomo. Para simplificar el modelo se supone que las energías de los orbitales
atómicos $\varepsilon_{n}^{\prime}$ puede ser distinta pero las correspondientes
funciones de onda $\mid n\rangle$ son todas iguales, donde el índice $n$
indica el átomo $n$-simo. Se propone una solución a la ecuación de
Schrödinger para un electrón en el cristal en forma de combinación lineal de
orbitales atómicos $\mid \Psi\rangle=\sum_n \psi_n \mid n\rangle$.

Para obtener las ecuaciones para las amplitudes $\psi_n$ partimos del
Hamiltoniano de un único electrón en el potencial del cristal

$$\hat{H}=\hat{T}+\sum_{m}V_m=\hat{H_l}+\sum_{m\neq l}V_m$$

donde $\hat{T}$ es el operador de energía cinética y $\hat{H_l}$
es el Hamiltoniano que describe un único electrón en el potencial $V_l$
del átomo $l$-simo. Por tanto 

$$\hat{H_l} \mid l \rangle = \varepsilon_{l}^{\prime} \mid l \rangle \ .$$

La ecuación de Schrödinger para el electrón en el potencial del cristal
es 

$$\hat{H} \mid \Psi \rangle = E \mid \Psi \rangle\ .$$ 

Introduciendo la solución propuesta obtenemos que

$$
\sum_l \psi_l \left[\hat{H_l}-E+\sum_{m\neq l}V_m\right] \mid l \rangle=
\sum_l \psi_l \left[\varepsilon_l^{\prime}-E+\sum_{m\neq l}V_m\right] 
\mid l \rangle= 0\ .
$$

A continuación multiplicamos por la izquierda por $\langle n\mid$ y suponemos
que el solapamiento de orbitales de átomos diferentes es despreciable, es
decir, $\langle n\mid l \rangle=\delta_{nl}$. 

Con ello llegamos a que

$$
E\psi_n=\varepsilon_n^{\prime}\psi_n+\sum_l \psi_l \sum_{m\neq l} 
\langle n \mid V_m \mid l \rangle\ .
$$

La aproximación de enlace fuerte considera exclusivamente los elementos de
matriz $\langle n\mid V_m \mid l \rangle\ $ que incluyan sitios que son vecinos
próximos y desprecia los restantes. Teniendo en cuenta que en la ecuación
anterior $n$ es arbitrario pero fijo y los índices $m$ y $l$ son diferentes,
sólo tenemos las siguientes posibilidades, $m=n\ $ y $l=n\pm 1\ $ o bien $m=n\pm 1\ $
y $l=n\ $. Por tanto, dentro de esta aproximación, resulta

\begin{eqnarray*}
E\psi_n&=&\Big[ \varepsilon_n^{\prime}+
\langle n\mid V_{n+1} \mid n \rangle+\langle n\mid V_{n-1} \mid n \rangle
\Big] \psi_n \\
&+&\langle n\mid V_n \mid n+1 \rangle\psi_{n+1}+
\langle n\mid V_n \mid n-1 \rangle\psi_{n-1}
\end{eqnarray*}

Como se ha supuesto que los orbitales atómicos tienen la misma forma
$\langle n\mid V_n \mid n+1 \rangle=\langle n\mid V_n \mid n-1 \rangle
\equiv -J$

Finalmente, definiendo

$$\varepsilon_n \equiv \varepsilon_n^{\prime}
+\langle n\mid V_{n+1}\mid n \rangle+\langle n\mid V_{n-1} \mid n \rangle$$

llegamos a que 

$$
E\psi_n=\varepsilon_n\psi_n-J\left(\psi_{n+1}+\psi_{n-1}\right)\ .
$$





## Cadena atómica uniforme infinita

Si la cadena es uniforme todas las energías de los orbitales son iguales $\varepsilon_n=\varepsilon\ $. En tal caso, de acuerdo con el teorema de Bloch para potenciales periodicos (infinitos), la ecuación anterior admite una solución de la forma

$$
\psi_n=\exp(iKn)
$$

Tras sustituirla en la ecuación para las amplitudes obtenemos la relación de dispersión, que relaciona la energía del estado con el correspondiente momento cristalino $K$

$$
E_K=\varepsilon-2J\cos K
$$

El valor de $K$ está comprendido entre $-\pi$ y $\pi\ $, correspondiente a la primera zona de Brillouin. La relación de dispersión obtenida describe una única banda, que se extiende desde $-2J\ $ hasta $+2J\ $. 

## Cadena atómica uniforme finita

La ecuación para las amplitudes es general y permite estudiar cadenas 
atómicas para las que las energías $\varepsilon_n$ son diferentes. Sin embargo,
cuando la cadenas es uniforme todas las energías de los orbitales son
iguales. Denotando por $\varepsilon$ la energía de cada orbital, la ecuación
anterior se puede escribir como

$$
E\psi_n=\varepsilon\psi_n-J\left(\psi_{n+1}+\psi_{n-1}\right)\ .
$$

Si la cadena tiene $N$ átomos, entonces las condiciones de contorno 
apropiadas para resolver la ecuación para las amplitudes son 
$\psi_0=\psi_{N+1}=0\ $. 

La ecuación para las amplitudes se puede resolver mediante la siguiente
función

$$
\psi_n = \sin(n\theta)
$$ 

donde por el momento hemos omitido el factor de normalización. Introduciendo
la solución propuesta en la ecuación y empleando la identidad trigonométrica

$$\sin[(n+1)\theta]+\sin[(n-1)\theta]=2\cos(\theta)\sin (n\theta)$$

encontramos que

$$
E=\varepsilon -2J \cos(\theta)\ .
$$

Puesto que se tienen que cumplir las condiciones de contorno, no todos los
valores del parámetro $\theta\ $ son posibles. La solución propuesta cumple
efectivamente que $\psi_0=0\ $ pero para que también cumpla que $\psi_{N+1}=0\ $
debe ocurrir que 

$$\sin[(N+1)\theta]=0\ ,$$ 

es decir 

$$\theta=k\pi/(N+1)\ ,$$ 

donde $k$ es un número entero. Podemos elegir 

$$k=1,2,\cdots,N$$ 

puesto que cualquier
otro valor de $k$ fuera de este intervalo conduce a una amplitud $\psi_n\ $
correspondiente a un valor $k$ en dicho intervalo. 

En definitiva, las amplides normalizadas son

$$
\psi_n=\sqrt{\frac{1}{N+1}}\ \sin\left(\frac{nk\pi}{N+1}\right) 
$$

y las correspondientes energías son

$$
E_k=\varepsilon-2J\cos\left(\frac{k\pi}{N+1}\right) \qquad
k=1,2,\cdots,N\ .
$$

Notemos que las energías obtenidas son similares a las correspondientes a una cadena infinita sin más que hacer la identificación

$$
K \rightarrow \frac{k\pi}{N+1}
$$

## Cadena diatómica uniforme

No es difícil generalizar el cálculo anterior al caso de una cadena 
diatómica, donde las energías de los orbitales atómicos pueden ser
$\varepsilon_A$ y $\varepsilon_B$. Si los átomos A ocupan las posiciones 
pares y los átomos B las impares, las ecuaciones para las amplitudes
serán

$$E\psi_{2n}  =\varepsilon_A \psi_{2n}-2J\left(\psi_{2n-1}+\psi_{2n+1}\right)$$

$$E\psi_{2n+1}=\varepsilon_B \psi_{2n+1}-2J\left(\psi_{2n}+\psi_{2n+2}\right)$$

En este caso proponemos como solución

$$\psi_{2n}  =A \sin(2n\theta)\qquad \psi_{2n+1}=B \sin[(2n+1)\theta]$$

donde $A$ y $B$ son amplitudes a determinar. Insertando las soluciones 
propuestas en las ecuaciones de movimiento llegamos a que

$$
(E-\varepsilon_A)A + 2J\cos(\theta)  B =  0
$$
$$
2J\cos(\theta)  A+(E-\varepsilon_B)B   =  0
$$

Anulando el determinante de este sistema de ecuaciones encontramos la
siguiente relación de dispersión

$$
E=\varepsilon \pm \sqrt{\Delta^2+4J^2\cos^2(\theta)}
$$

siendo 

$$\varepsilon=(\varepsilon_A+\varepsilon_B)/2 \qquad \Delta=(\varepsilon_A-\varepsilon_B)/2$$ 
0
Al igual que antes, las condiciones de
contorno conducen a los valores permitidos de $\theta$. Consideremos que tenemos
un número par de átomos $N\ $. Entonces el hipotético nodo $0$ correspondería a
una amplitud de la forma $A\sin(2n\theta)\ $ con $n=0\ $, por lo que cumple la
condición de contorno $\psi_0=0\ $. Por otra parte, el nodo $N+1\ $ tendría asociada
una amplitud $\psi_{N+1}=B\sin[(2n+1)\theta]\  $ con $2n=N\ $ y como la amplitud debe anularse,
entonces $\theta=\pi k/(N+1)\ ,$ que es la misma condición que obtuvimos en la red
atómica uniforme. En definitiva

$$
E_k=\varepsilon \pm \sqrt{\Delta^2+4J^2\cos^2\left(\frac{k\pi}{N+1}\right)}
\qquad
k=1,2,\cdots,N\ .
$$


## Cadena atómica desordenada

Algunas de las propiedades de los sólidos desordenados se pueden entender
empleando la aproximación de enlace fuerte cuando las energías de los órbitales
toman valores aleatorios. El comportamiento de los electrones en los sólidos
desordenados difiere de lo que ocurre en los sólidos cristalinos.  En
particular, el desorden causa la localización espacial de algunos de esos
electrones.  Desde un punto de vista teórico, la mecánica cuántica
de los electrones en sólidos carentes de simetría de traslación se
complica puesto que no se puede utilizar el teorema de Bloch.

Cuando el número de átomos no es elevado, los estados electrónicos se
pueden obtener mediante diagonalización del Hamiltoniano. Para ello, nos damos
cuenta de que las ecuaciones para las amplitudes se pueden escribir de
forma matricial como

$$
\mathcal{H} \vec{\psi} = E \vec{\psi}
$$

donde $\vec{\psi}$ es un vector cuyas $N$ componentes son las amplitudes
$\psi_n$ y la matriz $\mathcal{H}$ es tridiagonal de orden $N\times N$. Los elementos
de la diagonal son $\varepsilon_n$ y los restantes elementos no nulos son
iguales a $-J$

$$
\mathcal{H} = \left(
\\begin{array}{ccccccc}
\epsilon_1 & -J & 0 & 0   & \cdots & 0 & 0\\\\
-J &  \epsilon_2 & -J & 0 & \cdots & 0 & 0\\\\
0 & -J &  \epsilon_3 & -J & \cdots & 0 & 0\\\\
\vdots & \vdots &\vdots &  \vdots & \mbox{} & \vdots & \vdots\\\\
0 & 0& 0 & 0 & \cdots &  \epsilon_{N-1} & -J \\\\
0 & 0& 0 & 0 & \cdots & -J& \epsilon_N
\\end{array}
\right)
$$

Habitualmente las energías de los orbitales se eligen a partir de una
distribución de probabilidad con un cierto valor medio, que se puede tomar
igual a cero sin pérdida de generalidad, y una cierta anchura $\Delta\ $. En el caso
particular de una distribución uniforme

$$
\mathcal{P} (\epsilon_n) = \frac{1}{\Delta}\times \\left\\{
\\begin{array}{cc}
1 & \mathrm{si\ } |\epsilon_n| \leq \Delta/2 \\\\
0 & \mathrm{si\ } |\epsilon_n| > \Delta /2
\\end{array}
\\right.
$$

La forma de proceder es generar una secuencia de $N$ números aleatorios de acuerdo con la distribución uniforme anterior, y posteriormente diagonalizar
la matriz tridiagonal $\mathcal{H}\ $, para lo que existen algoritmos muy rápidos y eficientes. De esta manera podemos obtener las autoenergías y los
correspondientes autoestados $\vec{\psi}\ $.

## Estados electrónicos asociados a impurezas

Los sólidos reales presentan muchos tipos de defectos. Los defectos de menor tamaño son los denominados defectos puntuales, como pueden ser las vacantes, los átomos intersticiales y las impurezas. En el modelo que hemos presentado aquí podemos simular la presencia de una única impureza en un cristal suponiendo que todas las energías de los orbitales son iguales excepto una, que denominaremos $\varepsilon^{\prime}\ $. En esta situación podemos volver a obtener los estados electrónicos de la cadena mediante diagonalización directa, como hicimos en el apartado anterior. La única diferencia es que ahora los elementos de la diagonal de la matriz $\mathcal{H}$ son todos iguales a $\varepsilon$ excepto uno de ellos, que será igual a $\varepsilon^{\prime}\ $. 

### Ejercicio Cadena uniforme

#### Comparar la banda de energía de la cadena infinita y de la cadena finita con diferente numero de sitios. Variar el valor de N entre 0 y 20.

```python
%pylab inline
import sympy as sym
K = sym.symbols('K') 
eps=0.0 # energía de los orbitales
J=1.0 # energía del acoplamiento electrónico a primeros vecinos (se usará como unidad de energía)
sym.plot(eps-2*J*sym.cos(K),(K,-pi,pi),xlabel='ka',title='E(k)/J Relacion de Dispersion Analitica')

N=5 # número de átomos en la cadena
k=linspace(-N,N,2*N+1)
Ek=eps-2*J*cos(k*pi/(N+1))

scatter(k*pi/(N+1),Ek)
xlim([-4.,4.])
ylim([-2.1,2.1])
xlabel('ka',fontsize=20)
title('E(k)/J Relacion de Dispersion Numerica', fontsize=14)
```

### Ejercicio Cadena Diatómica

#### Estudiar la evolución del gap de energía cuando se modifica la diferencia de energía entre los orbitales de los átomos A y B.

```python
epsA=1. # energía de los orbitales del átomo A
difeps=0.5# diferencia energética de los orbitales de los átomos A y B
epsB=epsA+difeps # energía de los orbitales del átomo B
J=1.0 # energía del acoplamiento electrónico a primeros vecinos (se usará como unidad de energía)
epsC=(epsA+epsB)/2
delta=(epsA-epsB)/2.

N=300 # número de átomos en la cadena
k=linspace(-N,N,2*N+1)
Ediat1=epsC+sqrt(delta*delta + 4.*J*J*cos(k*pi/(N+1))*cos(k*pi/(N+1)))
Ediat2=epsC-sqrt(delta*delta + 4.*J*J*cos(k*pi/(N+1))*cos(k*pi/(N+1)))

plot(k*pi/(N+1),Ediat1)
plot(k*pi/(N+1),Ediat2) 
xlabel('ka',fontsize=20)
ylabel('E(k)/J',fontsize=20)
title('Relacion De Dispersion Cadena Diatomica',fontsize=14)
```

### Ejercicio Cadena Desordenada

#### Estudiar la localización de Anderson observando la región de localización de los autoestados en función del grado de desorden en la cadena (delta)

```python
%pylab inline
N=100 # número de átomos en la cadena
J=1.0 # energía del acoplamiento electrónico a primeros vecinos (se usará como unidad de energía)
delta=0.5 # anchura de la distribución de probabilidad uniforme de desorden

# Definición de la matriz del Hamiltoniano del sistema desordenado
DiagElements=random.uniform(-delta/2,delta/2,N)
M=identity(N)*DiagElements
for j in range (0,N):
    for k in range (0,N):
        if (j==(k+1) or j==(k-1)):
            M[j,k]=-J

# Solución del problema de autovalores y autovectores del Hamiltoniano
e,w = eig(M)
E=empty(N)
W=empty([N,N])
index=argsort(e)
for j in range(0,N):
    E[j]=e[index[j]]
for j in range(0, N):
    for k in range(0,N):
        W[j,k]=w[k,index[j]]

# Los autovalores y autovectores del Hamiltoniano ordenados según la energía se almacenas en E y W 
site=linspace(1,N,N)
level=1 # Nivel de autovalor y autovector que se dibuja
        

plot(site, absolute(W[level]))
xlabel('n', fontsize=20)
ylabel('$|\psi|^2$',fontsize=20)
title('Densidad de Probabilidad del Autoestado (numero=level)')
```

### Ejercicio Cadena Finita Uniforme con una Impureza
