#  Electronic states in a one-dimensional atomic lattice

Some transition metal oxides such as K$_{0.3}$MoO$_{3}$ behave as one-dimensional conductors. A simple model allows us to capture the underlying physics of the electron dynamics in one-dimensional conductors. A single orbital per atom is tunnel coupled to nearest neighbor orbitals within the tigh-binding approximation. To keep the model as simple as possible, we will assume that the energy of the orbitals $\varepsilon_{n}^{\prime}$ can dependent on the atomic site but the corresponding wave functions $\mid n\rangle$ are the same. Here $n$ labels the atomic site. 

Solution to the Schrödinger equation is taken as a linear combination of atomic orbitals $\mid \Psi\rangle=\sum_n \psi_n \mid n\rangle$. The equation for the amplitudes $\psi_n$ are obtained from the one electron Hamiltonian in the lattice potential

$$\hat{H}=\hat{T}+\sum_{m}V_m=\hat{H_l}+\sum_{m\neq l}V_m$$

where $\hat{T}$ is the kinetic energy operator and $\hat{H_l}$ is the one electron Hamiltonian in the potential of the $l$th atom $V_l$. Thus

$$\hat{H_l} \mid l \rangle = \varepsilon_{l}^{\prime} \mid l \rangle \ .$$

The Schrödinger equation for the electron in the lattice potential is

$$\hat{H} \mid \Psi \rangle = E \mid \Psi \rangle\ .$$ 

Inserting the ansatz introduced above in the Schrödinger equation yields

$$
\sum_l \psi_l \left[\hat{H_l}-E+\sum_{m\neq l}V_m\right] \mid l \rangle=
\sum_l \psi_l \left[\varepsilon_l^{\prime}-E+\sum_{m\neq l}V_m\right] 
\mid l \rangle= 0\ .
$$

Multipling by $\langle n\mid$ and using the approximation $\langle n\mid l \rangle=\delta_{nl}\ $ leads to

$$
E\psi_n=\varepsilon_n^{\prime}\psi_n+\sum_l \psi_l \sum_{m\neq l} 
\langle n \mid V_m \mid l \rangle\ .
$$

The tight-binding approximation keeps only those matrix elements $\langle n\mid V_m \mid l \rangle$ involving nearest neighbor sites and disregards the rest. In the above equation the site index $n$ is fixed but otherwise arbitrary while $m$ and $l$ are different. Therefore, we will keep two sets of matrix elements involving 

$$m=n\ ,\quad l=n\pm 1\ ,$$

or 

$$m=n\pm 1\ ,\quad l=n\ ,$$

and obtain

\begin{eqnarray*}
E\psi_n&=&\Big[ \varepsilon_n^{\prime}+
\langle n\mid V_{n+1} \mid n \rangle+\langle n\mid V_{n-1} \mid n \rangle
\Big] \psi_n \\
&+&\langle n\mid V_n \mid n+1 \rangle\psi_{n+1}+
\langle n\mid V_n \mid n-1 \rangle\psi_{n-1}
\end{eqnarray*}

Since we assumed that the wave function of the atomic orbitals are the same,
then $\langle n\mid V_n \mid n+1 \rangle=\langle n\mid V_n \mid n-1 \rangle
\equiv -J$

Finally we define

$$\varepsilon_n \equiv \varepsilon_n^{\prime}
+\langle n\mid V_{n+1}\mid n \rangle+\langle n\mid V_{n-1} \mid n \rangle$$

and conclude that

$$
E\psi_n=\varepsilon_n\psi_n-J\left(\psi_{n+1}+\psi_{n-1}\right)\ .
$$


## Infinite monoatomic lattice

Let us discuss the simplest case when all orbital energies are the same $\varepsilon_n=\varepsilon\ $. In this scenario Bloch theorem ensures that the amplitudes are of the form

$$
\psi_n=\exp(iKn)\ ,
$$

apart from a normalization constant. Inserting it in the equation of motion for the amplitudes yields the dispersion relation

$$
E_K=\varepsilon-2J\cos K\ .
$$

The dimensionless crystal momentum lies within the first Brillouin zone that ranges from $-\pi$ to $\pi\ $. The dispersion relation describes a single band from $-2J$ up to $+2J$.. 

## Finite monoatomic lattice

The equation of motion for the amplitudes holds in the case of finite lattices, namely when the number of atoms $N$ is finite and boundary conditions read $\psi_0=\psi_{N+1}=0\ $. The equation for the amplitudes is then solved by the following ansatz

$$
\psi_n = \sin(n\theta)
$$ 

apart from a normalization constant. Inserting this ansatz into the equation for the amplitudes and taking into account that 

$$\sin[(n+1)\theta]+\sin[(n-1)\theta]=2\cos(\theta)\sin (n\theta)$$

we find that

$$
E=\varepsilon -2J \cos(\theta)\ .
$$

Allowed values of the parameter $\theta\ $ are obtained from the boundary conditions. The proposed ansatz already satisfies that $\psi_0=0\ $. To meet the condition $\psi_{N+1}=0\ $ one gets


$$\sin[(N+1)\theta]=0\ ,$$ 

or equivalenty

$$\theta=k\pi/(N+1)\ ,$$ 

where $k$ is an integer number. We can take 

$$k=1,2,\cdots,N$$ 

since any other $k$ value outside this set leads to the same solution spanned by the set. 

Normalized amplitudes then read

$$
\psi_n=\sqrt{\frac{1}{N+1}}\ \sin\left(\frac{nk\pi}{N+1}\right) 
$$

and the corresponding energies are

$$
E_k=\varepsilon-2J\cos\left(\frac{k\pi}{N+1}\right) \qquad
k=1,2,\cdots,N\ .
$$


## Finite diatomic lattice

Generalization of the above results to the case of a diatomic lattices is straightforward. In the diatomic lattice the site energies take on two values $\varepsilon_A$ and $\varepsilon_B$. Assuming that A (B) atoms are labelled by an even (odd) site index, the equation of motion for the amplitude can be cast in the form

$$E\psi_{2n}  =\varepsilon_A \psi_{2n}-2J\left(\psi_{2n-1}+\psi_{2n+1}\right)$$

$$E\psi_{2n+1}=\varepsilon_B \psi_{2n+1}-2J\left(\psi_{2n}+\psi_{2n+2}\right)$$

The ansatz is taken now as

$$\psi_{2n}  =A \sin(2n\theta)\qquad \psi_{2n+1}=B \sin[(2n+1)\theta]$$

where $A$ and $B$ are constants to be determined. Using the equation for the amplitudes we obtain

$$
(E-\varepsilon_A)A + 2J\cos(\theta)  B =  0
$$
$$
2J\cos(\theta)  A+(E-\varepsilon_B)B   =  0
$$.

Making the determinant to vanish we can find the dispersion relation


$$
E=\varepsilon \pm \sqrt{\Delta^2+4J^2\cos^2(\theta)}
$$

where

$$\varepsilon=(\varepsilon_A+\varepsilon_B)/2 \qquad \Delta=(\varepsilon_A-\varepsilon_B)/2$$ 

Boundary conditions determine the allowed values of $\theta$. Let $N$ be the (even) number of atoms in the diatomic lattice. Then, the fictitious site $0$ would have an amplitude $A\sin(2n\theta)\ $ with $n=0\ $, and then $\psi_0=0\ $ as expected. On the other side, the amplitude of the fictitious site $N+1\ $ would be $\psi_{N+1}=B\sin[(2n+1)\theta]\  $ with $2n=N\ $. This amplitude vanishes if $\theta=\pi k/(N+1)\ $. Hence

$$
E_k=\varepsilon \pm \sqrt{\Delta^2+4J^2\cos^2\left(\frac{k\pi}{N+1}\right)}
\qquad
k=1,2,\cdots,N\ .
$$


## Disordered lattice

Some of the properties of disordered solids can be understood by using the tight-binding approach when the energies of the orbitals take on random values. The behavior of electrons in disordered solids differs from what occurs in crystalline solids. In particular, disorder causes the spatial localization of some of those electrons. From a theoretical point of view, quantum mechanics of electrons in solids with broken translation symmetry is
complicated since Bloch theorem can not be used.

When the number of atoms is not high, the electronic states can be obtained by Hamiltonian diagonalization. For this, we note that the equations for the amplitudes can be cast in matrix form as

$$
\mathcal{H} \vec{\psi} = E \vec{\psi}
$$

where $\vec{\psi}$ is a column vector whose $N$ components are the amplitudes
$\psi_n$ and the matrix $\mathcal{H}$ is tridiagonal of size $N\times N$. 
Diagonal elements are $\varepsilon_n$ and the rest of non-vanishing elements are equal to $-J$

$$
\mathcal{H} = \left(
\begin{array}{ccccccc}
\epsilon_1 & -J & 0 & 0   & \cdots & 0 & 0\\\\
-J &  \epsilon_2 & -J & 0 & \cdots & 0 & 0\\\\
0 & -J &  \epsilon_3 & -J & \cdots & 0 & 0\\\\
\vdots & \vdots &\vdots &  \vdots & \mbox{} & \vdots & \vdots\\\\
0 & 0& 0 & 0 & \cdots &  \epsilon_{N-1} & -J \\\\
0 & 0& 0 & 0 & \cdots & -J& \epsilon_N
\end{array}
\right)
$$

Usually the energies of the orbitals are chosen from a probability distribution with a certain mean value, which can be taken equal to zero without loss of generality, and a certain width $\Delta \ $. In the particular case of a uniform distribution

$$
\mathcal{P} (\epsilon_n) = \frac{1}{\Delta}\times \left\{
\begin{array}{cc}
1 & \mathrm{si\ } |\epsilon_n| \leq \Delta/2 \\\\
0 & \mathrm{si\ } |\epsilon_n| > \Delta /2
\end{array}
\right.
$$

The procedure is to generate a sequence of $N$ random numbers according to the previous uniform distribution, and then diagonalize the tridiagonal matrix $\mathcal{H}\ $, for which there are very fast and efficient algorithms. In this way we can obtain the eigenenergies and the
corresponding eigenstates $\vec{\psi}\ $.

## Electronic states in an impurity

Real solids have many types of defects. The smaller defects are the so-called point defects, such as vacancies, interstitial atoms and impurities. In the model we have presented here we can simulate the presence of a single impurity in a crystal assuming that all the energies of the orbitals are equal except one, which we will call $\varepsilon^{\prime}\ $. In this situation we can obtain the electronic states of the lattice by direct diagonalization, as we did in the previous section. The only difference is that now the diagonal elements of the matrix $\mathcal{H}\ $ are all equal to $\varepsilon$ except one of them, which will be equal to $\varepsilon^{\prime}\ $.


### Exercise. Uniform lattice.

#### Obtain and compare the energy bands of the infinity and finite lattices. The number of atoms $N$ of the finite lattice is set between $5$ and $20$.

```python
%pylab inline
import sympy as sym
K = sym.symbols('K') 
eps=0.0 # Energy of the atomic orbitals
J=1.0 # Tunnel coupling constant, used as a unit of energy
sym.plot(eps-2*J*sym.cos(K),(K,-pi,pi),xlabel='K',title='E(k)/J Analytical dispersion relation')

N=5 # Number of atoms in the finite lattice
k=linspace(-N,N,2*N+1)
Ek=eps-2*J*cos(k*pi/(N+1))

scatter(k*pi/(N+1),Ek)
xlim([-4.,4.])
ylim([-2.1,2.1])
xlabel('k',fontsize=20)
title('E(k)/J Numerical dispersion relation', fontsize=14)
```

### Exercise. Diatomic lattice.

#### Study the evolution of the energy gap as a function of the energy difference of the two atomic orbitals.

```python
epsA=1. # Energy of the A atom
difeps=0.5# Energy difference of the atomic orbitals A and B
epsB=epsA+difeps # Energy of the B atoms
J=1.0 # Tunnel coupling constant, used as a unit of energy
epsC=(epsA+epsB)/2
delta=(epsA-epsB)/2.

N=300 # Even number of atoms in the lattice
k=linspace(-N,N,2*N+1)
Ediat1=epsC+sqrt(delta*delta + 4.*J*J*cos(k*pi/(N+1))*cos(k*pi/(N+1)))
Ediat2=epsC-sqrt(delta*delta + 4.*J*J*cos(k*pi/(N+1))*cos(k*pi/(N+1)))

plot(k*pi/(N+1),Ediat1)
plot(k*pi/(N+1),Ediat2) 
xlabel('k',fontsize=20)
ylabel('E(k)/J',fontsize=20)
title('Dispersion relation in the diatomic lattice',fontsize=14)
```

### Exercise. Disordered lattice.

#### Study Anderson localization of the electron states, observing the region of localization of the eigenstates according to the magnitude of disorder in the lattice (delta)

```python
%pylab inline
N=100 # Number of atoms in the lattice
J=1.0 # Tunnel coupling constant, used as a unit of energy
delta=0.5 # Magnitude of disorder

# System matrix Hamiltonian
DiagElements=random.uniform(-delta/2,delta/2,N)
M=identity(N)*DiagElements
for j in range (0,N):
    for k in range (0,N):
        if (j==(k+1) or j==(k-1)):
            M[j,k]=-J

# Diagonalization of the matrix Hamiltonian
e,w = eig(M)
E=empty(N)
W=empty([N,N])
index=argsort(e)
for j in range(0,N):
    E[j]=e[index[j]]
for j in range(0, N):
    for k in range(0,N):
        W[j,k]=w[k,index[j]]

# Sorted eigenvelues and eingenvectors
site=linspace(1,N,N)
level=1 # Level to be plotted
        

plot(site, absolute(W[level]))
xlabel('n', fontsize=20)
ylabel('$|\psi|^2$',fontsize=20)
title('Probability densisty of the level 1')
```

### Exercise. Finite lattice with a single impurity.

#### Study the eigenstate and the eigenenergy of the lattice associated with an impurity. How is the dependence on the eigenenergy of the impurity energy?

```python
%pylab inline
N=50 # Number of atoms in the lattice
J=1.0 # Tunnel coupling constant, used as a unit of energy
impurity=20 # Position of the impurity

# System matrix Hamiltonian
DiagElements=zeros(N) # All atomic energies are set to zero.
DiagElements[impurity]=2.0 # Impurity energy
M=identity(N)*DiagElements
for j in range (0,N):
    for k in range (0,N):
        if (j==(k+1) or j==(k-1)):
            M[j,k]=-J

# Diagonalization of the matrix Hamiltonian
e,w = eig(M)
E=empty(N)
W=empty([N,N])
index=argsort(e)
for j in range(0,N):
    E[j]=e[index[j]]

for j in range(0, N):
    for k in range(0,N):
        W[j,k]=w[k,index[j]]

# Sorted eigenvelues and eingenvectors
site=linspace(1,N,N)
level=49


figsize(14,4)
subplot(121)
plot(site, absolute(W[level]))
xlabel('n', fontsize=20)
ylabel('$|\psi|^2$',fontsize=20)
title('Probability densisty of the level 49')
subplot(122)
scatter(site,E)
xlabel('n', fontsize=20)
ylabel('E(k)/J',fontsize=20)
title('Energy levels in the lattice with one impurity')
```
