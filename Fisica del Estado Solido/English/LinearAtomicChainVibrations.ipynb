Consider an infinitely long one-dimensional atomic chain. If $V(|x_n-x_\ell |)\ $ is the interaction potential between the atoms $n$ and $\ell$, which are in the positions $x_n$ and $x_\ell$, respectively, then the classic Hamiltonian of the lattice is

$$
H=\sum_n \frac{1}{2}m\dot{x}_{n}^{2} + \sum_n \sum_{\ell \neq n} \frac{1}{2} V(|x_n-x_\ell|)
$$

We assume that the atomic displacements with respect to the equilibrium position are small. Therefore, we can write that

$$
x_{n} = na + u_{n} 
\qquad
|u_{n}| \ll a
$$

where $a$ is the lattice parameter. This approach allows us to perform a Taylor expansion of potential around the equilibrum position and disregard the cubic and higher order terms. For brevity of notation we define $z_{n\ell} = u_{n} -u_{\ell}\ $ so that

$$
V(|x_n-x_\ell|) \simeq V(|n-\ell|a)+z_{n\ell}\frac{\partial V}{\partial z_{n\ell}} + \frac{1}{2}z_{n\ell}^2\frac{\partial^2 V}{\partial z_{n\ell}^2}
$$

where all the derivatives are evaluated in $z_{n\ell} = 0\ $. The zero-order term is the potential energy of the lattice in equilibrium while the linear term in the displacements with respect to the equilibrium corresponds to the force between the atoms in that position. The total force on an atom in the equilibrium position is zero, so we do not have to consider it. Thus

$$
H=\sum_n \frac{1}{2}m\dot{u}_{n}^{2} + H_0+\sum_n \sum_{\ell \neq n} \frac{1}{4} K(|n-\ell|)(u_{n}-u_{\ell})^2
$$

where $H_0$ is the energy in equilibrium and $K(| n- \ell |)$ is the second derivative of the interaction potential when the separation between the atoms is $|n-\ell|a\ $.

From now on we will consider only the interaction between nearest neighbors and next-nearest neighbors only. Let $K_1\equiv K(a)\ $ and $ K_2\equiv K(2a)\ $ so that the classical Hamiltonian would approximately be

$$
H=H_0+\sum_{n} \left\{\frac{1}{2}m\dot{u}_{n}^{2} +\frac{1}{4} K_1 \Big[\Big(u_n-u_{n+1}\Big)^2+\Big(u_n-u_{n-1}\Big)^2\Big]
+\frac{1}{4} K_2 \Big[\Big(u_n-u_{n+2}\Big)^2+\Big(u_n-u_{n-2}\Big)^2\Big]\right\}
$$

Equation of motion is then

$$
m\ddot{u}_n=K_1\Big(u_{n+1}+u_{n-1}-2u_n\Big)+K_2\Big(u_{n+2}+u_{n-2}-2u_n\Big)
$$


## Dispersion relation

To find the dispersion relation we propose a solution of the form

$$
u_n=Ae^{i(nka-\omega t)}
$$

and after substitution in the equation of motion one finds

$$
\omega^2 =4\omega_1^2\Big(\sin^2(ka/2)+\frac{K_2}{K_1}\sin^2(ka)\Big)
$$

where for brevity we have defined

$$
\omega_1^2=\frac{K_1}{m}
$$

## Density of modes

When periodic boundary conditions are imposed, it is found that the allowed values of $k$ are $k = 2\pi j/L\ $ where $j$ is an integer and $L$ is the length of the lattice. Therefore, the separation between two consecutive values of $k$ is $2\pi/L\ $, so that the density of modes in the space of moments is the inverse of that amount $N(k)= L/2\pi\ $. The density of modes will then be such that

$$
N(\omega)d\omega=N(k)dk=(L/2\pi)dk\ ,
$$ 

so that the density of modes per unit of length turns out to be

$$
n(\omega)\equiv \frac{1}{L}N(\omega)=\frac{1}{2\pi} \frac{1}{d\omega/dk}
$$

That is, we can obtain the density of modes per unit of length directly from the dispersion relation.

### Calculation and representation of the dispersion relation and density of vibration modes of a one-dimensional atomic lattice

#### Study how the dispersion relation and the density of modes vary for different values of the ratio $K_2/K_1$. Discuss what happens when $ K_2 \gg K_1 $

```python
%pylab inline
m=1. # Atomic mass
a=1. # Lattice parameter
K1=1. # Coupling constant between nearest neighbor atoms
K2=1.# Coupling constant between next-nearest neighbor atoms
k=linspace(-1.*pi,1.*pi,51)
N=size(k)
frec2=4.*(sin(k*a/2)**2+(K2/K1)*(sin(k*a))**2) # cuadrado de las frecuencias de vibracion permitidas en la red en unidades de sqrt(K1/m)
frec=sqrt(4.*(sin(k*a/2)**2+(K2/K1)*(sin(k*a))**2)) # frecuencias de vibracion permitidas en la red en unidades de sqrt(K1/m)

# Calculation of the density of vibration modes through the derivative of the dispersion relation
w=empty(N-1)
nw=empty(N-1)
for j in range (1,N):
    w[j-1]=(frec[j]+frec[j-1])/2
    nw[j-1]=absolute(((k[j]-k[j-1])/(frec[j]-frec[j-1])))

W=empty(N-1)
NW=empty(N-1)
index=argsort(w)
for j in range(0,N-1):
    W[j]=w[index[j]]
    NW[j]=nw[index[j]]

    
figsize(14,4)
subplot(121)

# Dispersion relation in the first Brillouin zone

plot(k,sqrt(frec2))
xlabel('ka',fontsize=20)
ylabel('$\omega/\omega_1$',fontsize=20)
title('Dispersion relation in the first Brillouin zone',fontsize=14)

# Density of modes 

ax2=subplot(122)
scatter(NW,W)
axis([0,20, 0, W[size(W)-1]+0.2]);
ylabel('$\omega/\omega_1$',fontsize=20)
xlabel('$D(\omega)$ [arb. units]',fontsize=20)
title('Density of modes',fontsize=20)

# Shaded area

b1=0
b2=amax(W)
verts = [(0,b1)] + list(zip(NW,W))+[(0,b2)]
polygon = Polygon(verts)
ax2.add_patch(polygon)


```
