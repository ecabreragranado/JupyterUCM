### <div class='alert'> Task 3. Analysis of the data. Linear adjustment of the spokes of the dark rings </div>

To analyze the results, indicate in the next code cell the number of rings measured and run the cell. Boxes will appear where we will include the real values of the radii of the dark rings obtained in the previous task in mm.

```python
#INDICATE THE NUMBER OF RINGS MEASURED

measured_rings = 3 ## Indicate here the number of rings measured


#DO NOT HERE TO TOUCH
##################

#from IPython.html import widgets
import IPython.html.widgets
from IPython.display import display

listwidgets = []
for i in range (measured_rings):
     listwidgets.append (widgets.FloatTextWidget (description = 'Actual Radii of the Ring' + str (i + 1), value = 1))
     display(listwidgets [i])
```

Using the radii of the dark rings we will graphically represent the radius squared based on the number of the dark ring. This representation should give us a linear dependence whose slope is the radius of curvature multiplied by the wavelength of the light used in the experiment, that is,

$$ \text {slope} = \lambda \text {R} $$

Since we have used white light to perform the experiment, we will use a central wavelength of the visible: $ \lambda $ = 550 nm.

In the next code cell the data is represented and the linear adjustment is performed to obtain the value of the slope (it is written in the figure). This value will have the units of the radii squared, that is, mm $ ^ 2 $.

```python
# DO NOT TOUCH ONLY EXECUTE
#################################################################################################### #####################
% pylab inline

values_radii = []
for i in range (measured_rings):
     values_radii.append (listwidgets [i] .value)

radii = array (values_radii) # The radii of the dark rings

m = linspace (1, size (radii), size (radii)) # Vector with the numbers of the rings
radii2 = radii * radii # Vector with the radii of the rings squared
a, b = polyfit (m, radii2,1) # Adjustment of data to a line where a is the slope and b the ordinate at the origin

plot (m, radii2, 'o', m, a * m + b, '-', lw = 2)
xlabel ('Ring number')
ylabel ('Radio $ ^ 2 $ (mm $ ^ 2 $)') # We write the names of the axes
te = "slope =% f mm $ ^ 2 $"% a; title (te); # The value of the slope is shown;
```
